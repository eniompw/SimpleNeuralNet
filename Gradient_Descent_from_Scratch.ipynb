```python
# Importing Libraries
import numpy as np

# Data y = 2x + 50
x = np.arange(100)		# 200 causes an overflow in training
y = 2 * x + 50
# Add Gaussian noise to data
noise = np.random.normal(0, 5, 100)
y = y + noise

# Print first 10 values
print("x y")
for i in range(10):
	print(x[i] , y[i])
```

```python
# Draw all Data points
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(x,y)
```

```python
weight = 0.1
bias = 0.01
y_pred = (weight * x) + bias

# Draw Data and Model
plt.figure(figsize=(10,10))
plt.plot(x,y_pred,'-r')
plt.scatter(x,y)
```

```python
diffx = [x[50], x[50]]
diffy = [y[50], weight * x[50] + bias]

# Draw Data, Model and Error
plt.figure(figsize=(10,10))
plt.plot(diffx, diffy, '-g')
plt.scatter(x,y)
plt.plot(x,y_pred,'-r')
```

```python
def mean_squared_error(y, y_pred):
	# Calculating the loss or cost
	cost = np.sum((y - y_pred) ** 2) / len(y)
	return cost
```

```python
learning_rate = 0.0001
costs = []

for i in range(50000):
  # Calculating the current cost
  current_cost = mean_squared_error(y, y_pred)
  costs.append(current_cost)

  # Calculating the gradients
  n = len(y)
  weight_derivative = -(2/n) * sum(x * (y - y_pred))
  bias_derivative = -(2/n) * sum(y - y_pred)

  # Updating weights and bias
  weight = weight - (learning_rate * weight_derivative)
  bias = bias - (learning_rate * bias_derivative)
  y_pred = (weight * x) + bias

print("weight:" , weight)
print("bias:" , bias)
```

```python
# Visualising the learning
plt.figure(figsize=(10,10))
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.plot(costs[10:])
```

```python
# Final Model vs Data
plt.figure(figsize=(10,10))
y_pred = (weight * x) + bias
plt.plot(x,y_pred,'-r')
plt.scatter(x,y)
```
